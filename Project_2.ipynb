```python
import pandas as pd
import requests
from sqlalchemy import create_engine
from config import username, pwd
```

```python
school_data = pd.read_csv("Public_Schools.csv")
school_data.head()
```

```python
income_data = pd.read_csv("IRSIncomeByZipCode.csv")
income_data.head()
```

```python
income_data = income_data.drop(columns=["Adjusted gross income (AGI)","Number of returns with total income","Total income amount","Number of returns with taxable income","Avg taxable income"])
income_data.head()
```

```python
#Renaming ZipCODE as Zip
income_data = income_data.rename(columns={"ZIPCODE":"ZIP"})
income_data.head()
```

```python
school_data.columns
```

```python
school_data = school_data.drop(columns=["NCESID","ZIP4","STATUS","COUNTYFIPS","NAICS_CODE","NAICS_DESC","SOURCEDATE",
                         "VAL_METHOD","VAL_DATE","DISTRICTID","FT_TEACHER"])
```

```python
school_data['TYPE'].replace('1', 'Regular School')
```

```python
school_data.head()
```

```python
school_data.dtypes
```

```python
school_data['TYPE'] = school_data['TYPE'].apply(str)
```

```python
school_data['TYPE'] = school_data['TYPE'].replace('1', 'Regular School')
```

```python
school_data.head()
```

```python
school_data['TYPE'] = school_data['TYPE'].replace('2', 'Special Education')
```

```python
school_data['TYPE'] = school_data['TYPE'].replace('3', 'Vocational')
```

```python
school_data['TYPE'] = school_data['TYPE'].replace('4', 'Other/Alternative')
```

```python
school_data['TYPE'].unique()
```

```python
# Merging the data based on Zipcodes
project_2 = school_data.merge(income_data,how='left',on="ZIP")
project_2
```

```python
#Removing unwanted columns from the merged dataset
project_2 = project_2.drop(columns=["SHELTER_ID","STATE_y","X","Y","SOURCE","Number of returns"])
project_2
```

```python
project_2 = project_2.rename(columns={"STATE_x":"STATE"})
project_2
```

```python
# Converting dataframe to store as CSV file
project_data = project_2.to_csv("project_data.csv")
```

```python
# Load the DataFrame into Postgres Database, Please ensure to run the table schemata sql file before running this part of code
#Create Engine
engine = create_engine(f'postgresql://{username}:{pwd}@localhost/Project_2')
```

```python
# Insert into project_data table in SQL
project_2.to_sql(name='project_data',con=engine,if_exists='append',index=False)
```
